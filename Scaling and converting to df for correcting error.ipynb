```python
import pandas as pd
import numpy as np

df = pd.read_csv('Final PF-12 and Injection.csv',index_col=0,parse_dates=True)
```

```python
to_drop =["BORE_GAS_VOL",'AVG_DOWNHOLE_PRESSURE','AVG_DOWNHOLE_TEMPERATURE','NPD_WELL_BORE_CODE', 'NPD_WELL_BORE_NAME','Water Injection F-4', 'Injection Hours F-5','Injection Hours F-4','Water Injection F-5','BORE_OIL_VOL','BORE_WAT_VOL']
X = df.drop(to_drop,axis = 1)
Y = df['BORE_OIL_VOL']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.1,shuffle=False)
```

```python
X_train.head()
```

```python
y_train_df = pd.DataFrame(y_train)
y_test_df = pd.DataFrame(y_test)
```

```python
y_train_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled_values = scaler.fit_transform(X_train)

X_train_scaled = pd.DataFrame(data = X_train_scaled_values , index = X_train.index , columns = X_train.columns)


X_test_scaled_values = scaler.transform(X_test)

X_test_scaled = pd.DataFrame(data = X_test_scaled_values , index = X_test.index , columns = X_test.columns)
```

```python
y_train_scaled_values = scaler.fit_transform(y_train_df)

y_train_scaled = pd.DataFrame(data = y_train_scaled_values , index = y_train_df.index , columns = y_train_df.columns)


y_test_scaled_values = scaler.transform(y_test_df)

y_test_scaled = pd.DataFrame(data = y_test_scaled_values , index = y_test_df.index , columns = y_test_df.columns)
```
